```python
df_escolas = df2['id_escola'].value_counts()
df_escolas = df_escolas.reset_index()
df_escolas.columns = ['id_escola', 'Quant. Estudantes']
df_escolas
```

```python
df_escolas['Quant Estudantes com Risco Baixo'] = df_escolas.loc[:,[ 'E_ESC_R1','E_PROF_R1','E_FAM_R1', 'E_EST_R1', 'E_COM_R1', ]].sum(axis=1)

df_escolas['Quant Estudantes com Risco Médio'] = df_escolas.loc[:,[ 'E_ESC_R2','E_PROF_R2','E_FAM_R2', 'E_EST_R2', 'E_COM_R2', ]].sum(axis=1)

df_escolas['Quant Estudantes com Risco Alto'] = df_escolas.loc[:,[ 'E_ESC_R3','E_PROF_R3','E_FAM_R3', 'E_EST_R3', 'E_COM_R3', ]].sum(axis=1)

df_escolas
```

```python
df_escolas['Perc Risco Baixo'] = ((df_escolas['Quant Estudantes com Risco Baixo'] / df_escolas['Quant. Estudantes']) *100)/5
df_escolas['Perc Risco Médio'] = ((df_escolas['Quant Estudantes com Risco Médio'] / df_escolas['Quant. Estudantes']) *100)/5
df_escolas['Perc Risco Alto'] = ((df_escolas['Quant Estudantes com Risco Alto'] / df_escolas['Quant. Estudantes']) *100)/5

df_escolas
```

```python
def map_escolas():
    fig = px.scatter_mapbox(df_escolas, 
                            title="Escolas BR",
                            zoom=3, center={"lat": -11.44, "lon": -52.01},
                            lat="Latitude", lon="Longitude",
                            color="Perc Risco Alto",
                            hover_name="Nome da Escola", 
                            hover_data=[
                                'id_escola',
 'Quant. Estudantes',
 'Nome da Escola',
 'Cidade',
 'Estado',
 'Localizacao',
 'Localidade Diferenciada',
 'Modo Ensino',
 'Resticao de Atendimento',
 'Cod escola INEP',
 'Categoria Administrativa',
 'Telefone',
 'Dependencia Administrativa',
 'Categoria Escola',
 'Conveniada Poder Publico',
 'Regulamentacao Conselho',
 'Porte',
 'Etapas e Modalidade Ensino',
 'Outras Ofertas',
 'Cod Regiao',
 'Micro Regiao',
 'Meso Regiao',
 'Regiao',
 'Cod Estado',
 'Quant de Estudantes',
 'Latitude',
 'Longitude',
 'Quant Estudantes com Risco Alto',
 'Quant Estudantes com Risco Baixo',
 'Quant Estudantes com Risco Médio',
 'Perc Risco Baixo',
 'Perc Risco Médio',
 'Perc Risco Alto'
                            ]
                            
                                )
    #fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(mapbox_style="carto-positron")
    fig.update_traces(marker=dict(sizemin=2))
    return fig

fig = map_escolas()
#export_fig_html(fig, "locais_votacao_2022.html")
fig.show()
#fig.write_html("maps_escola.html")
```
