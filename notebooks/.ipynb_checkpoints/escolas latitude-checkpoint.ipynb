```python
df_for = ['E_ESCV', 'E_PROFV', 'E_FAMV', 'E_COMV', 'E_ESTV']
df_escolas['Nome da Escola'] = ''

for index, row in df_escolas.iterrows():
    #print(row['id_escola'])
    #print(index)
    busca = "id_escola == " + str(row['id_escola'])
    esc = df2.query(busca)
    #print(esc['escola'].values[0])
    df_escolas.loc[index, 'Nome da Escola'] = esc['escola'].values[0]
    
    df_temp = df[df['Código INEP']==row['id_escola']]
    #df_temp
    df_escolas.loc[index, 'Latitude'] = df_temp['Latitude'].values[0]
    df_escolas.loc[index, 'Longitude'] = df_temp['Longitude'].values[0]
    
    df_escolas.loc[index, 'Cidade'] = esc['municipio'].values[0]
    df_escolas.loc[index, 'Estado'] = esc['uf'].values[0]
    df_escolas.loc[index, 'Localizacao'] = esc['localizacao'].values[0]
    df_escolas.loc[index, 'Localidade Diferenciada'] = esc['localidade_diferenciada'].values[0]
    df_escolas.loc[index, 'Modo Ensino'] = esc['modo_ensino'].values[0]
    df_escolas.loc[index, 'Resticao de Atendimento'] = esc['resticao_de_atendimento'].values[0]
    df_escolas.loc[index, 'Cod escola INEP'] = esc['cod_escola'].values[0]
    df_escolas.loc[index, 'Categoria Administrativa'] = esc['categoria_administrativa'].values[0]
    df_escolas.loc[index, 'Telefone'] = esc['telefone'].values[0]
    df_escolas.loc[index, 'Dependencia Administrativa'] = esc['dependencia_administrativa'].values[0]
    df_escolas.loc[index, 'Categoria Escola'] = esc['categoria_escola'].values[0]
    df_escolas.loc[index, 'Conveniada Poder Publico'] = esc['conveniada_poder_publico'].values[0]
    df_escolas.loc[index, 'Regulamentacao Conselho'] = esc['regulamentacao_conselho'].values[0]
    df_escolas.loc[index, 'Porte'] = esc['porte'].values[0]
    df_escolas.loc[index, 'Etapas e Modalidade Ensino'] = esc['etapas_e_modalidade_ensino'].values[0]
    df_escolas.loc[index, 'Outras Ofertas'] = esc['outras_ofertas'].values[0]
    df_escolas.loc[index, 'Cod Regiao'] = esc['cod_regiao'].values[0]
    df_escolas.loc[index, 'Micro Regiao'] = esc['micro'].values[0]
    df_escolas.loc[index, 'Meso Regiao'] = esc['meso'].values[0]
    df_escolas.loc[index, 'Regiao'] = esc['regiao'].values[0]
    df_escolas.loc[index, 'regiao'] = esc['regiao'].values[0]
    df_escolas.loc[index, 'Cod Estado'] = esc['cod_estado'].values[0]
    

    
    df_escolas.loc[index, 'Quant de Estudantes'] = len(esc)
    df_escolas.loc[index, 'E_ESCV'] = esc['E_ESCV'].mean()
    df_escolas.loc[index, 'E_PROFV'] = esc['E_PROFV'].mean()
    df_escolas.loc[index, 'E_FAMV'] = esc['E_FAMV'].mean()
    df_escolas.loc[index, 'E_COMV'] = esc['E_COMV'].mean()
    df_escolas.loc[index, 'E_ESTV'] = esc['E_ESTV'].mean()
    
    df_escolas.loc[index, 'E_ESC_R1'] = len(esc[esc['E_ESCC']=='1'])
    df_escolas.loc[index, 'E_ESC_R2'] = len(esc[esc['E_ESCC']=='2'])
    df_escolas.loc[index, 'E_ESC_R3'] = len(esc[esc['E_ESCC']=='3'])
    
    df_escolas.loc[index, 'E_PROF_R1'] = len(esc[esc['E_PROFC']=='1'])
    df_escolas.loc[index, 'E_PROF_R2'] = len(esc[esc['E_PROFC']=='2'])
    df_escolas.loc[index, 'E_PROF_R3'] = len(esc[esc['E_PROFC']=='3'])
    
    df_escolas.loc[index, 'E_FAM_R1'] = len(esc[esc['E_FAMC']=='1'])
    df_escolas.loc[index, 'E_FAM_R2'] = len(esc[esc['E_FAMC']=='2'])
    df_escolas.loc[index, 'E_FAM_R3'] = len(esc[esc['E_FAMC']=='3'])
    
    df_escolas.loc[index, 'E_COM_R1'] = len(esc[esc['E_COMC']=='1'])
    df_escolas.loc[index, 'E_COM_R2'] = len(esc[esc['E_COMC']=='2'])
    df_escolas.loc[index, 'E_COM_R3'] = len(esc[esc['E_COMC']=='3'])
    
    df_escolas.loc[index, 'E_EST_R1'] = len(esc[esc['E_ESTC']=='1'])
    df_escolas.loc[index, 'E_EST_R2'] = len(esc[esc['E_ESTC']=='2'])
    df_escolas.loc[index, 'E_EST_R3'] = len(esc[esc['E_ESTC']=='3'])
    
    
    #row[index, 'nomesses'] = esc['escola'].mean()
    #row.loc[index, 'E_ESCV'] = esc['E_ESCV'].mean()
    #row[index, 'E_PROFV'] = esc['E_PROFV'].mean()
    #row[index, 'E_FAMV'] = esc['E_FAMV'].mean()
    #row[index, 'E_COMV'] = esc['E_COMV'].mean()
    #row[index, 'E_ESTV'] = esc['E_ESTV'].mean()
    #print(esc)
    
df_escolas
```

```python
df_escolas['Quant Estudantes com Risco Baixo'] = df_escolas.loc[:,[ 'E_ESC_R1','E_PROF_R1','E_FAM_R1', 'E_EST_R1', 'E_COM_R1', ]].sum(axis=1)

df_escolas['Quant Estudantes com Risco Médio'] = df_escolas.loc[:,[ 'E_ESC_R2','E_PROF_R2','E_FAM_R2', 'E_EST_R2', 'E_COM_R2', ]].sum(axis=1)

df_escolas['Quant Estudantes com Risco Alto'] = df_escolas.loc[:,[ 'E_ESC_R3','E_PROF_R3','E_FAM_R3', 'E_EST_R3', 'E_COM_R3', ]].sum(axis=1)

df_escolas
```

```python
df_escolas['Perc Risco Baixo'] = ((df_escolas['Quant Estudantes com Risco Baixo'] / df_escolas['Quant. Estudantes']) *100)/5
df_escolas['Perc Risco Médio'] = ((df_escolas['Quant Estudantes com Risco Médio'] / df_escolas['Quant. Estudantes']) *100)/5
df_escolas['Perc Risco Alto'] = ((df_escolas['Quant Estudantes com Risco Alto'] / df_escolas['Quant. Estudantes']) *100)/5

df_escolas
```

```python
list(df_escolas)
```

```python
def map_escolas():
    fig = px.scatter_mapbox(df_escolas, 
                            title="Escolas BR",
                            zoom=3, center={"lat": -11.44, "lon": -52.01},
                            lat="Latitude", lon="Longitude",
                            color="Perc Risco Alto",
                            hover_name="Nome da Escola", 
                            hover_data=[
                                'id_escola',
 'Quant. Estudantes',
 'Nome da Escola',
 'Cidade',
 'Estado',
 'Localizacao',
 'Localidade Diferenciada',
 'Modo Ensino',
 'Resticao de Atendimento',
 'Cod escola INEP',
 'Categoria Administrativa',
 'Telefone',
 'Dependencia Administrativa',
 'Categoria Escola',
 'Conveniada Poder Publico',
 'Regulamentacao Conselho',
 'Porte',
 'Etapas e Modalidade Ensino',
 'Outras Ofertas',
 'Cod Regiao',
 'Micro Regiao',
 'Meso Regiao',
 'Regiao',
 'Cod Estado',
 'Quant de Estudantes',
 'Latitude',
 'Longitude',
 'Quant Estudantes com Risco Alto',
 'Quant Estudantes com Risco Baixo',
 'Quant Estudantes com Risco Médio',
 'Perc Risco Baixo',
 'Perc Risco Médio',
 'Perc Risco Alto'
                            ]
                            
                                )
    #fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(mapbox_style="carto-positron")
    fig.update_traces(marker=dict(sizemin=2))
    return fig

fig = map_escolas()
#export_fig_html(fig, "locais_votacao_2022.html")
fig.show()
fig.write_html("maps_escola.html")
```
